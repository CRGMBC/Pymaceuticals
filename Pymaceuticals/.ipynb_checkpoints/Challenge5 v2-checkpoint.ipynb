This study compared the performance of 10 drug regimens for the treatment of squamous cell carcinoma (SCC), a form of skin cancer, on mice over 45 days.  Thie purpose was to compare the performance of Pymaceuticals’ drug of interest, Capomulin, against the other treatment regimens.  Ramicane, Infubinol, and Ceftamin were the most promising treatment regimens along with the interest drug Capomulin.  Statistical measures such as quartiles and interquartile range (IQR) were determined to identify potential outliers and only the drug Infubinol showed any significant outlier.  It should be noted that there was a strong positive correlation between the weight of the mouse and the average tumour volume.  The use of vizualisation techniques assists in the communication of the statistucal results for decision making processes within Pymaceuticals, Inc.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df=pd.merge(study_results,mouse_metadata, how='left', on='Mouse ID')
# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
mousenumber = merge_df['Mouse ID'].unique()
count_mouseno = len(mousenumber)
count_mouseno
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mice = merge_df[merge_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_mouse_ids = duplicate_mice['Mouse ID'].unique()
duplicate_mouse_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_merge_df = merge_df[~merge_df['Mouse ID'].isin(duplicate_mouse_ids)]
clean_merge_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleanmouseno = clean_merge_df['Mouse ID'].unique()
count_cleanmouseno = len(cleanmouseno)
count_cleanmouseno
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

grouped = clean_merge_df.groupby("Drug Regimen")
summary_table = grouped["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
summary_table = summary_table.rename(columns={"mean": "Mean Tumor Volume", "median": "Median Tumor Volume", "var": "Tumor Volume Variance", "std": "Tumor Volume Std. Dev.", "sem": "Tumor Volume Std. Err."})
summary_table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

summary_table = grouped["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Start by getting the last (greatest) timepoint for each mouse - use maximum timepoint
last_regimen = clean_merge_df.groupby('Mouse ID')['Timepoint'].max()
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_record_df = pd.merge(clean_merge_df, last_regimen, on=['Mouse ID', 'Timepoint'], how='inner')
last_record_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds
for regimen in treatment_regimens:
    final_tumor_volume = last_record_df.loc[last_record_df['Drug Regimen'] == regimen, 'Tumor Volume (mm3)']
    tumor_volume_data.append(final_tumor_volume)

    outliers = []

for i, regimen in enumerate(treatment_regimens):
    quartiles = tumor_volume_data[i].quantile([0.25, 0.5, 0.75])
    
    lower_bound = quartiles[0.25] - 1.5 * (quartiles[0.75] - quartiles[0.25])
    upper_bound = quartiles[0.75] + 1.5 * (quartiles[0.75] - quartiles[0.25])
    
    treatment_outliers = tumor_volume_data[i][(tumor_volume_data[i] < lower_bound) | (tumor_volume_data[i] > upper_bound)]
    outliers.append(treatment_outliers)
   
    print(f"{regimen}'s potential outliers: {treatment_outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig, ax = plt.subplots()
flierprops = dict(marker='o', markerfacecolor='red', markersize=10, linestyle='none', markeredgecolor='black')
ax.boxplot(tumor_volume_data, labels=treatment_regimens, flierprops=flierprops)

ax.set_xlabel("Treatment")
ax.set_ylabel('Final Tumor Volume (mm3)')
ax.set_title('Final tumor volume for mice per treatment group')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin (example used mouse l509)
mousetoplot = 'l509'
capomulin_data = clean_merge_df[(clean_merge_df['Drug Regimen'] == 'Capomulin') & (clean_merge_df['Mouse ID'] == mousetoplot)]

x = capomulin_data['Timepoint']
y = capomulin_data['Tumor Volume (mm3)']

plt.plot(x, y, marker=' ')

plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capomulin treatment of mouse l509')

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
entire_capomulin_data = clean_merge_df[clean_merge_df['Drug Regimen'] == 'Capomulin']
average_tumor_volume = entire_capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
merged_data = pd.merge(average_tumor_volume, entire_capomulin_data[['Mouse ID', 'Weight (g)']], on='Mouse ID', how='left')

x = merged_data['Weight (g)']
y = merged_data['Tumor Volume (mm3)']

plt.scatter(x, y)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title(f'Mouse weight vs average observed tumor volume for Capomulin')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
entire_capomulin_data = clean_merge_df[clean_merge_df['Drug Regimen'] == 'Capomulin']
average_tumor_volume = entire_capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = entire_capomulin_data.groupby('Mouse ID')['Weight (g)'].first()

regression_data = pd.DataFrame({'Average Tumor Volume': average_tumor_volume, 'Mouse Weight': mouse_weight})
correlation_coefficient = regression_data['Mouse Weight'].corr(regression_data['Average Tumor Volume'])
pe_slope, pe_int, pe_r, pe_p, pe_std_err = st.linregress(mouse_weight, average_tumor_volume)
pe_fit = pe_slope * mouse_weight + pe_int

print(f"The correlation between mouse weight and the average tumor volume is {correlation_coefficient:.2f}")
#print(f"Linear Regression Model: y = {pe_slope:.2f}x + {pe_int:.2f}")

plt.scatter(mouse_weight,average_tumor_volume)
plt.plot(mouse_weight,pe_fit,"r--")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title(f'mouse weight vs average observed tumor volume for Capomulin')
plt.show()
```
